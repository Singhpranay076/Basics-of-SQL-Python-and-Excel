```python
#With functions, you write the code once and reuse it:

def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50))
```

```python

#In Python, a function is defined using the def keyword, followed by a function name and parentheses:
#The code inside the function must be indented. Python uses indentation to define code blocks.

def check_login(username, password):  
    return username == "admin" and password == "1234"

username = input("Enter username: ")
password = input("Enter password: ")

if check_login(username, password):
    print("Login successful")
    print("Access granted")
else:
    print("Invalid login")
```

```python
def check_login(username, password):
    return username == "admin" and password == "1234"

while True:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if check_login(username, password):
        print("Login successful")
        print("Access granted")
        break
    else:
        print("Invalid login. Try again.")
```

```python
balance = 1000

action = input("Enter action (deposit/withdraw/check): ")

if action == "deposit":
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print("New balance:", balance)

elif action == "withdraw":
    amount = float(input("Enter amount to withdraw: "))
    balance -= amount
    print("New balance:", balance)

elif action == "check":
    print("Your balance is:", balance)

else:
    print("Invalid action")
```

```python
balance = 1000

action = input("Enter action (deposit/withdraw/check): ")

if action == "deposit":
    amount = float(input("Enter amount to deposit: "))
    balance += amount                              #+= is a shortcut assignment operator in Python. balance = balance + amount
    print("New balance:", balance)

elif action == "withdraw":
    amount = float(input("Enter amount to withdraw: "))
    
    if balance - amount >= 1000:
        balance -= amount
        print("New balance:", balance)
    else:
        print("Withdrawal denied. Minimum balance must remain 1000.")

elif action == "check":
    print("Your balance is:", balance)

else:
    print("Invalid action")
```
