```python
import pandas as pd
import numpy as np
```

```python
ord = pd.read_csv("orders.csv")
prd = pd.read_csv("products.csv")
```

```python
# MERGING using JOINs

com_tab = pd.merge(ord, prd, on = "product_id", how = "left")
print(com_tab)
```

```python
result= com_tab.groupby(["region","product_id", "product_category"])["revenue"].sum().reset_index(name = "Total_revenue")
print(result)
```

```python
# Applying Conditions

result["Order_type"] = result["Total_revenue"].apply(
    lambda x: "High" if x >= 1200 else "Low" 
)    

# Sorting 

result = result.sort_values(by="Total_revenue", ascending=False)

print(result)
```

```python
com_tab.to_csv("Day_3.csv", index=False)   # To Extract Result as .csv file
```
