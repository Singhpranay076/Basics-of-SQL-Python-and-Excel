```python
#Create List

fruits = ["Apple", "Banana", "Cherry", "Dragon", "Elderberry","Feijoa", "Guava"]
print(fruits)
print(fruits[3])    # square bracket is for the list---- 3--Index

#change data in the list

fruits[3] = "Dragon fruits"
print(fruits)

#append-- add single item to the list at the very end.

fruits.append("Frooti")
print(fruits)

#extend--- add list to existing list

fruits.extend(["Huckle berry", "Indian Prunes"])
print(fruits)
```

```python
tasks = []                    #This list will store all tasks.

for i in range(3):
    task = input("Enter a task: ")
    tasks.append(task)       #append() adds each task to the list.

print("Your To-Do List:")
for t in tasks:          #The loop goes through each task and prints it.
    print("-", t)
```

```python
inventory = {
    "apples": 10,
    "bananas": 5,               #This is a dictionary:      key → item name     value → quantity
    "oranges": 8
}                          

item = input("Enter item name: ")

if item in inventory:
    print("Quantity:", inventory[item])
else:
    print("Item not found")
```

```python
tasks = []

for i in range(3):
    name = input("Enter task name: ")
    
    task = {
        "task": name,
        "done": False
    }
    
    tasks.append(task)

print("\nYour To-Do List:")

for t in tasks:
    print(t["task"], "- Done:", t["done"])
```

```python
playlist = []

for i in range(3):
    song = input("Enter song name: ")
    artist = input("Enter artist name: ")
    
    music = {
        "song": song,
        "artist": artist
    }
    
    playlist.append(music)

print("\nYour Mixtape:")

for m in playlist:
    print(m["song"], "by", m["artist"])
    
```

```python
expenses = []

for i in range(3):
    item = input("Enter expense name: ")
    amount = float(input("Enter amount: "))
    
    expense = {                       #Each expense is stored as a dictionary.
        "item": item,
        "amount": amount
    }
    
    expenses.append(expense)         #append() means add an item to the end of a list.

total = 0

print("\nBudget List:")

for e in expenses:
    print(e["item"], "-", e["amount"])
    total += e["amount"]       #This adds each expense amount to the total.

print("Total spending:", total)
```
