```python
import pandas

mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pandas.DataFrame(mydataset)                  # dataframe ---- A DataFrame is a two-dimensional table made up of multiple Series.
 
print(myvar)
```

```python
import pandas as pd                                  # pd as alias 

a = [1, 7, 2]                                        # list creation

myvar = pd.Series(a, index = ["x", "y", "z"])        # Index manipulation---- Uses of Series 
                                                     # A Series is essentially a one-dimensional labeled array.

print(myvar)
print(myvar["y"])
```
