```python
# Rollercoaster

print("Welcome to the rollercoaster")
height= int(input("What is your height in cm ?"))
bill = 0

if height>=120:                                                            # nested if-else statement
   
    print("Yay!! You can Enjoy the ride")
    age=int(input("What's your age "))
    
    if age>=18 and age<=44:
       bill= 10
       print(f"You have to pay Adult rate -- {bill}")
   
    elif age<5:                                                              # Greedy Rollercaoster owner got beaten here !!
       bill= 0
       print(f"Impossible and free !!!")
    elif age>= 45 and age <=55:
        bill = 0
        print(f"Exclusive Discount -- 0")
   
    else:
       bill= 0 
       print(f"You can pay a discounted rate -- {bill} ")
    
    photo = input(f"Do you want a photo of a roller coaster. Press y for ES and n for NO")
    
    if photo.lower()== "y":  # add extra €3 
       bill += 3
    print(f"Total payable bill is = € {bill}")

else:
   print("Sorry!! Come back again")
```

```python
## even or odd

num= int(input(print("Enter a number you want to check !")))
r= num % 2         #modulus 
print(r)
if r==0:
    print("Its an Even Number")
else:
    print("Its a Odd Number")
    
 
```

```python
#bmi machine

height= float(input("What is your height in m ?"))
weight= float(input("What is your weight in kgs ?"))
bmi = float(weight/(height ** 2))

if bmi < 18.5:
    print("underweight")
elif bmi < 25:
    print("normal weight")
else:
    print("overweight")
```

```python
print("\\@ Welcome to Pranay's Veg Pizza Corner \\@")

size= input("Enter size of Pizza S/M/L ")
spicy= input("Do you want it be extra spicy -- Y for Yes or N for No")   # Extra 50 for small 60 for M/L
cheese=input("Do you want to add extra cheesy layer --  Y for Yes or N for No ") # Extra 100 for small and 120 for M/L
bill = 0

if size.lower()== "s":
    bill += 100
elif size.lower()=="m":
    bill += 200
elif size.lower()=="l":
    bill += 300
else:
    print("wrong input")

if spicy.lower()== "y":
   if size.lower()=="s":
        bill += 50
   else:
        bill += 60

if cheese.lower()== "y":
    if size.lower()=="s":
        bill += 100
    else:
        bill += 120

print(f"Your bill is - {bill}")
```
