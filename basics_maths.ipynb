```python
import datetime

x = datetime.datetime.now()
print(x)
```

```python
#The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

import datetime

x = datetime.datetime(2026, 3, 12)

print(x.strftime("%B"))
print(x.strftime("%c"))
print(x.strftime("%X"))
```

```python
#Built-in Math Functions
```

```python
#The min() and max() functions can be used to find the lowest or highest value in an iterable:

x = min(5, 10, 25)
y = max(5, 10, 25)

print(x)
print(y)
```

```python
#The abs() function returns the absolute (positive) value of the specified number:

x = abs(-7.25)

print(x)
```

```python
#The pow(x, y) function returns the value of x to the power of y (xy).
x = pow(1.01, 365)

print(x)
```

```python
import math

x = math.sqrt(64)

print(x)
```

```python
import math

x = math.ceil(1.4)
y = math.floor(1.4)

print(x) # returns 2
print(y) # returns 1
```

```python
import math

x = math.pi

print(x)
```
