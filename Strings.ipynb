```python
 # Variables got updated as moves down the code line

glass1 = "milk"
glass2 = "juice"

temp = glass1
glass1 = glass2
glass2 = temp

print(glass1)  # juice
print(glass2)  # milk
```

```python
glass1 = "milk"
glass2 = "juice"

glass1, glass2 = glass2, glass1 #One line code

print(glass1)
print(glass2)
```
