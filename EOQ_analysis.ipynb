```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# ==================== PARAMETERS (change these anytime) ====================
D = 1000   # Annual Demand (units/year)
S = 10     # Ordering cost per order ($)
H = 2      # Holding cost per unit per year ($)

# Calculate EOQ
EOQ = math.sqrt((2 * D * S) / H)
print(f"EOQ = {EOQ:.0f} units")

# Generate Order Quantity (Q) values from 10 to 300
Q = np.arange(10, 301, 10)

# Calculate the three costs
ordering_cost = (D / Q) * S
holding_cost  = (Q / 2) * H
total_cost    = ordering_cost + holding_cost
```

```python
# ====================== PLOT THE GRAPH ======================
plt.figure(figsize=(10, 6))  # size of the graph

plt.plot(Q, ordering_cost, label='Total Ordering Cost', color='blue', marker='o')
plt.plot(Q, holding_cost,  label='Total Holding Cost',  color='green', marker='o')
plt.plot(Q, total_cost,    label='Total Inventory Cost', color='red', marker='o', linewidth=2.5)

# Mark the EOQ point with a vertical dashed line
plt.axvline(x=EOQ, color='purple', linestyle='--', linewidth=2, label=f'EOQ = {EOQ:.0f} (Minimum Cost)')

# Titles and labels
plt.title('EOQ Model: Ordering vs Holding Cost Trade-off\n(D=1000, S=10, H=2)', fontsize=14)
plt.xlabel('Order Quantity (Q)')
plt.ylabel('Annual Cost ($)')
plt.legend()
plt.grid(True, alpha=0.3)
```
