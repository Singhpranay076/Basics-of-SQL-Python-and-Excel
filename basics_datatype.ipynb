```python
# List
numbers = [1, 2, 3]
numbers[0] = 10
print(numbers)   # [10, 2, 3]             #List: Mutable → you can change, add, or remove elements after creation.
                                          #shopping_list = ["milk", "bread", "eggs"]

# Tuple
numbers = (1, 2, 3)
numbers[0] = 10   # Error               #Tuple: Immutable → once created, you cannot modify it.
                                        #coordinates = (28.6139, 77.2090)
```

```python
a = {1,2,3}                       #A set is a mutable collection of unique elements.
b = {3,4,5}

print(a.union(b))        # {1,2,3,4,5}      #Union combines all unique elements from both sets.
print(a.intersection(b)) # {3}              #Intersection returns only the common elements present in both sets.
print(a.difference(b))   # {1,2}            #Difference returns elements that exist in the first set but not the second.
```

```python
numbers = {1, 2, 3}             #A set is a mutable collection of unique elements.
                                #Unordered (no indexing)
                                #No duplicate values 
                                #Elements can be added or removed

numbers.add(4)
numbers.remove(2)

print(numbers)
```
