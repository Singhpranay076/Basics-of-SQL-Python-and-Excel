```python
import pandas as pd    # Import pandas library
import numpy as np
```

```python
df = pd.read_csv("day_1.csv")      # To read file and import

print(df.head(5))
print(df.isnull().sum())        # Check for null rows
```

```python
Total_revenue = df["revenue"].sum()
print("Total revenue of all products is" , Total_revenue)  # Total Revenue
```

```python
df.groupby("region")["revenue"].sum()       # Total revenue after dividing it by Regions 
```

```python
#  Order Count per Region
df.groupby("region")["revenue"].count() 

# Average Revenue per Order

df.groupby("region")["revenue"].mean().reset_index().round(2)
```

```python
df.groupby("region")["revenue"].agg(["count", "mean"]).reset_index().round(2)
```
