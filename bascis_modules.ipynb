```python
import random

symbols = ["🍒", "🍋", "⭐", "7"]

spin = [random.choice(symbols),
        random.choice(symbols),
        random.choice(symbols)]

print("Slot result:", spin)

if spin[0] == spin[1] == spin[2]:
    print("Jackpot! You win!")
else:
    print("Try again!")
```

```python
import random

symbols = ["🍒", "🍋", "⭐", "7"]

wins = 0

for i in range(3):
    spin = [random.choice(symbols),
            random.choice(symbols),
            random.choice(symbols)]

    print("Spin", i+1, ":", " | ".join(spin))

    if spin[0] == spin[1] == spin[2]:
        print("Win!\n")
        wins += 1
    else:
        print("Lose\n")

if wins >= 2:
    print("You win the best of 3 game!")
else:
    print("You lose the best of 3 game.")
```

```python
import random

symbols = ["🍒", "🍋", "⭐", "7"]

wins = 0

for i in range(3):
    input("Press Enter to spin the wheel...")

    spin = [random.choice(symbols),
            random.choice(symbols),
            random.choice(symbols)]

    print("Spin", i+1, ":", " | ".join(spin))

    if spin[0] == spin[1] == spin[2]:
        print("Win!\n")
        wins += 1
    else:
        print("Lose\n")

if wins >= 2:
    print("You win the best of 3 game!")
else:
    print("You lose the best of 3 game.")
```

```python
solar_system = {
    "mercury": 58,
    "venus": 108,
    "earth": 150,
    "mars": 228
}

planet = input("Enter planet name: ").lower()

if planet in solar_system:
    print("Distance from Sun:", solar_system[planet], "million km")
else:
    print("Planet not found")
```
