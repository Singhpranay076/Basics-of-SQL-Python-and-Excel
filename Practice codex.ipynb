```python
import random

n = int(input("How many times do you want to flip the coin? "))

for i in range(n):
    result = random.choice(["Heads", "Tails"])
    print("Flip", i+1, ":", result)
    
```

```python
marks = int(input("Enter your marks: "))

if marks < 0 or marks > 100:
    print("Invalid marks")
elif marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
ph = float(input("Enter the pH value: "))

if ph < 7:
    print("The solution is Acidic")
elif ph == 7:
    print("The solution is Neutral")
else:
    print("The solution is Basic (Alkaline)")
    
```

```python
wind_speed = int(input("Enter wind speed (km/h): "))

if wind_speed < 63:
    print("Not a cyclone")
elif wind_speed <= 88:
    print("Cyclonic Storm")
elif wind_speed <= 117:
    print("Severe Cyclonic Storm")
elif wind_speed <= 165:
    print("Very Severe Cyclonic Storm")
else:
    print("Super Cyclone")
```
