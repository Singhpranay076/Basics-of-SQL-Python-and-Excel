```python
# With Pyplot, you can use the bar() function to draw bar graphs:

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color ='red', width = 0.75)                       # Vertical  layout
#plt.barh(x,y,color ='DarkBlue', height = 0.75)                # Horizonal layout - For horizontal bars, use "height" instead of "width".

plt.show()
```
