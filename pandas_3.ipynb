```python
# Now fix the problem. 

''' 1. In this case, we have NULL Values in Calories
    2. We have 7 duplicates.    '''

import pandas as pd

df = pd.read_csv('data.csv')

# To fix NULL values in Calories--- we will replace NULL values with mean- 375 

df['Calories'] = df['Calories'].fillna(375)

# To fix Duplicates- we will remove them

df.drop_duplicates()

print(df)
print(df.info())

df.corr()                  #The corr() method calculates the relationship between each column in your data set.


df.to_csv("Cleandata.csv", index=False)            # Export cleaned file for further usage.
```
