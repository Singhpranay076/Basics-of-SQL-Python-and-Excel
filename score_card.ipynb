```python
score_card=[99,88,77,66,55,44,33,22,11]

#total = sum(score)
#print(total)

# using loop 

sum =0
for score in score_card:
   sum+= score
   #print(sum) 
print(sum)                       # use of indentation outside the for loop to print the result, not the loop
```
