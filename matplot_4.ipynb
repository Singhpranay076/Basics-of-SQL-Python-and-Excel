```python
#we are generating 250 random numbers from a normal (Gaussian) distribution where:

# 170 →   mean (μ) ---                The center of the data
# 10 →    standard deviation (σ)---   How spread out the values are
# 250 →   number of sample----        Total data points

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x, bins=20, color='skyblue', edgecolor='black')

plt.title("Distribution of Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")

plt.show()


''' You’ll see a bell-shaped curve (roughly), because:
    1. Most values are near 170
    2. Fewer values at extremes (150 or 190) '''
```

```python
# Pie Charts 
# With Pyplot, you can use the pie() function to draw pie charts:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([37, 25, 25, 15])      # start from 1st quadrant and anticlockwise from x-axis towards y-axis
mylabels = ["Apples", "Bananas", "Cherries", "Dragon"]

plt.pie(y, labels=mylabels, autopct='%1.1f%%')

plt.legend(
    title="Pranay's favourite:",              # loc= "center left" → anchor point of legend
    loc="center left",                        # bbox_to_anchor=(1, 0.5) → pushes legend outside to the right
    bbox_to_anchor=(1, 0.5))                  # 1 = right edge of plot ----- 0.5 = vertically centered

plt.show()

(''' Note:
    1. If your data doesn’t sum to 100, matplotlib still converts it into proportions. It is wrongly assume it must be percentages.
    2. autopct= '%1.1f%%' → show percentage with 1 decimal place
    3. % is a special character in formatting. So, to print an actual %, you need '%%'
    4. %1.1f- % → start of formatting:      
                                      a) 1.1 → formatting rule
                                      b) first 1 → minimum width (not very important here)
                                      c) .1 → 1 decimal place
                                      d)  f → float (decimal number)      ''')
```

```python
#Start the first wedge at 90 degrees:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([37, 25, 25, 15])      # start from 1st quadrant and anticlockwise from x-axis towards y-axis
mylabels = ["Apples", "Bananas", "Dragon", "Cherries"]
mycolors = ["Red", "Yellow", "Pink", "Green"]

plt.pie(y, labels=mylabels, startangle = 90,colors = mycolors)
plt.show()
```

```python
#  The explode parameter, if specified, and not None, must be an array with one value for each wedge.
#  Add a shadow to the pie chart by setting the shadows parameter to True:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([37, 25, 25, 15])      
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, startangle = 90, explode = myexplode, shadow = True, autopct='%1.1f%%')
plt.show()
           
```

```python
#Extracting Image for Presentation

import matplotlib.pyplot as plt
import numpy as np

y = np.array([37, 25, 25, 15])      
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

# Control size for presentation

plt.figure(figsize=(8, 5))               # (8, 5) means 8 inches wide, 5 inches tall 
                                        # Actual pixel size depends on DPI.. here 300 so, 8*300 and 5*300 

plt.pie(
    y, 
    labels=mylabels, 
    startangle=90, 
    explode=myexplode, 
    shadow=True, 
    autopct='%1.1f%%'
)

plt.title("Fruit Distribution")

# Save in PNG (for PPT)
plt.savefig("profit_chart.png", dpi=300, bbox_inches='tight')

# Save in PDF (for high-quality reports)
plt.savefig("profit_chart.pdf", bbox_inches='tight')

plt.show()
```
