```python
import pandas as pd
import numpy as np
```

```python
ord = pd.read_csv("orders.csv")
prd = pd.read_csv("products.csv")
```

```python
# LEFT JOIN FOR MERGING 

com_tab = pd.merge(ord,prd, on= "product_id", how="left")
print(com_tab)

## com_tab.to_csv("Day_2.csv", index=False)      #export/save file in csv format
```

```python
com_tab.groupby(["region","product_id", "product_category"])["revenue"].sum().reset_index()
```
