```python
#Data cleaning means fixing bad data in your data set.

(''''Bad data could be:

1. Empty cells
2. Data in wrong format
3. Wrong data
4. Duplicates           ''')
```

```python
# Empty Cells -- Empty cells can potentially give you a wrong result when you analyze data.

import pandas as pd

df = pd.read_csv('sample.csv')

print(df)
```

```python
# Remove rows -- 

import pandas as pd

df = pd.read_csv('sample.csv')

new_df = df.dropna()            # By default, the dropna() method returns a new DataFrame, and will not change the original.

#dropna(inplace = True)        

print(new_df.to_string())

#print(df)
'''  Note: Now, the >>  dropna(inplace = True)  << will NOT return a new DataFrame, 
but it will remove all rows containing NULL values from the original DataFrame. '''
```

```python
# Replace-  The >> fillna() << method allows us to replace empty cells with a value:

import pandas as pd

df = pd.read_csv('sample.csv')

                                                   
df['Calories'] = df['Calories'].fillna(250)
df['Date'] = df['Date'].fillna('Unknown')

print(df)
```

```python
## Wrong format    
''' Cells with data of wrong format can make it difficult, or even impossible, to analyze data.

To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.  '''

# Pandas has a >> to_datetime() << method for this:

import pandas as pd

df = pd.read_csv('sample.csv')

#convert all cells in the 'Date' column into dates.

df['Date'] = pd.to_datetime(df['Date'], format='mixed')

#fillna() method allows us to replace empty cells with a value

df['Calories'] = df['Calories'].fillna(250)

print(df)


```

```python
# To discover duplicates, we can use the >> duplicated() << method.

import pandas as pd

df = pd.read_csv('sample.csv')
print(df.duplicated())
```

```python
# Removing Duplicates
#To remove duplicates, we can use the >> drop_duplicates() << method.

import pandas as pd

df = pd.read_csv('sample.csv')
new_df = df.drop_duplicates()

print(new_df)
```
