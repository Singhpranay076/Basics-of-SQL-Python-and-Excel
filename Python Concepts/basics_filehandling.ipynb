```python
#The key function for working with files in Python is the open() function.

f =open("simple.txt")             #Python opens the file in read mode by default (r).
f =open("simple.txt","rt")
#print(f.read())                  #read() → reads the entire file.

print(f.readline())               #Read one line only
print(f.readline())

#print(f.read(5))

f.close()                       #Without this, forgetting f.close() can cause memory issues.
```

```python
f = open("sample1.txt", "w")          #w = write (overwrites file)
f.write("Hello Python\n")
f.write("File handling practice")
f.close()
```

```python
f = open("sample1.txt")

for line in f:
    print(line)

f.close()
```

```python
with open("sample1.txt", "r") as f:   # use with---- Python automatically closes the file.
    print(f.read())
```

```python
with open("sample1.txt", "a") as f:
    f.write("\nNew line added")
```
