```python
class myclass():
  def __len__(self):        #__len__() tells Python the length of an object, similar to how len() works with lists or strings.
    return 0                #__len__  →  "How many things are inside?" ----- self     →  "this specific object"
                            #In Python, self plays the role of “your / this object”.

myobj = myclass() 
print(bool(myobj))
```

```python
class Person:
    def show(self):
        print("Self is:", self)

p1 = Person()
p2 = Person()

p1.show()        #Python actually runs: Person.show(p1) ----self = p1
p2.show()
```

```python
class Student:
    def show(self):
        print("Inside function, self =", self)

s1 = Student()

print("Outside object =", s1)
s1.show()
```
