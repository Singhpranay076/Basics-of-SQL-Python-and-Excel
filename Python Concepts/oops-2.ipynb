```python

#Create a class with properties:

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Pranay", 28)

print(p1.name)
print(p1.age)
print(f"Welcome !! {p1.name} I can see you are almost {p1.age} years old")
```

```python
#Change the age property:

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Pranay", 28)
print(p1.age)

p1.age = 26
p1.name = "Aman"

print(p1.age)
print(p1.name)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Pranay", 28)

del p1.age

print(p1.name) # This works
#print(p1.age) # This would cause an error
```

```python
#Class property vs instance property:

class Person:
  species = "Human"           # Class property

  def __init__(self, name):
    self.name = name          # Instance property

p1 = Person("Pranay")
p2 = Person("Shanvika")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```
