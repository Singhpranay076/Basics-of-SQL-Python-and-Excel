```python
#Tip Calculator
print("Welcome to the Tip Calculator")
total= float(input("What was the total bill? €"))
temp = int(input("How much tip would you like to give? 10|20|30 %"))
if temp==10:
    tip= 0.1*total
elif temp==20:
    tip= 0.2*total
elif temp==30:
    tip= 0.3*total
else:
    print("We don't expect other inputs")
    tip=0
round(tip,2)       #round-off
final= float(total+tip)
ppl= int(input("How many people to split the bill ?"))
bill=float(final/ppl)
round(bill,2)

#print("Each person should pay :", round(bill,2))

print(f"Each person should pay = {bill} , Thanks for adding tip of = {tip}" )

```
