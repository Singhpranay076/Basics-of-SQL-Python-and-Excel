```python
import random

number = random.randint(1, 10)

guess = 0                  #A variable guess is created and set to 0.
                           #This is just an initial value so the loop can start.
                            #It is different from the secret number.


while guess != number:         #while creates a loop.
                                #The condition means: keep running the program as long as the guess is not equal to the number.
                                #When the guess becomes equal to number, the loop stops.
    guess = int(input("Guess a number between 1 and 10: "))
    
    if guess < number:
        print("Too low")
    elif guess > number:
        print("Too high")
    else:
        print("Correct! You guessed it.")
```
