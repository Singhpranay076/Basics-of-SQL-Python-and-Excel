```python
from datetime import datetime, timedelta

# 1. Get today's date and time
today = datetime.now()
print("Today is:", today)

# 2. Add 7 days (useful for "delivery in one week")
next_week = today + timedelta(days=7)
print("One week from now:", next_week)

# 3. Subtract 3 days
three_days_ago = today - timedelta(days=3)
print("Three days ago:", three_days_ago)

# Add / Subtract
one_week_later   = today + timedelta(weeks=1)      # or days=7
one_week_ago     = today - timedelta(weeks=2)

five_hours_later = today + timedelta(hours=5)
three_hours_ago  = today - timedelta(hours=3, minutes=30)   # you can mix

print("1 week later  :", one_week_later)
print("2 weeks ago   :", one_week_ago)
print("5 hours later :", five_hours_later)
print("3.5 hours ago :", three_hours_ago)
```

```python
# Add 3 months
three_months_later = today.replace(month=today.month + 3)
print("Three months later :",three_months_later )

# If it goes beyond December, it automatically becomes next year
# Example: Dec 2026 + 2 months = Feb 2027
```

```python
# Use timedelta → for days, weeks, hours, minutes
# Use relativedelta → for months, years (and also days/weeks if you want)

from datetime import datetime
from dateutil.relativedelta import relativedelta

today = datetime.now()

print("Today                  :", today.date())
print("Next week              :", (today + relativedelta(weeks=1)).date())
print("5 hours later          :", (today + relativedelta(hours=5)).strftime("%H:%M"))
print("Next month             :", (today + relativedelta(months=1)).date())
print("3 months ago           :", (today - relativedelta(months=3)).date())
print("1 year and 2 months ago:", (today - relativedelta(years=1, months=2)).date())
print("1 year and 2 months later:", (today + relativedelta(years=1, months=2)).date())
```
