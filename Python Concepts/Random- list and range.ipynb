```python
import random

fruits = ["Apple", "Banana", "Cherry", "Dragon", "Elderberry","Feijoa", "Guava",'Frooti', 'Huckle berry', 'Indian Prunes']

print(fruits)
print(len(fruits))   # number of items in the list
# 1st method

fav= random.choice(fruits)               # Random choices
print(f"Your favourite fruit is -- {fav}")

#2nd method

ran_ind = random.randint(0,9)
print(f"Today's fruit is --- {fruits[ran_ind]}")
```

```python
#adding two lists

fruits = ["Strawberries", "Nectarines", "Apples", "Grapes", "Peaches", "Cherries", "Pears"]
vegetables = ["Spinach", "Kale", "Tomatoes", "Celery", "Potatoes"]
 
dirty_dozen = [fruits, vegetables]

        #dirty_dozen[0] → fruits list   #dirty_dozen[1] → vegetables list

print(dirty_dozen[1][1])
print(dirty_dozen[1][2])
print(dirty_dozen[1][3])

print("----------------")
print(dirty_dozen[0][1])
print(dirty_dozen[0][2])
print(dirty_dozen[0][3])
```
