```python
stages = ('''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''')



import random

word_list = ["sanjay", "poonam", "pranay", "aman", "nayanshi","shanvika"]
lives = 6

chosen_word = random.choice(word_list)
print(chosen_word)

placeholder= ""
word_length=len(chosen_word)
for position in range(word_length):
    placeholder += " _ "

print("Words to guess : " + placeholder)

game_over = False


correct_letters =[]

while not game_over:

    guess= input("Guess the Sachan's Family member --- type letter : ").lower()
        #print(guess)
    
    if guess in correct_letters:
        print(f"You have already guessed it {guess}")
    
    display= ""
    for letter in chosen_word:
        if letter == guess:
            display += letter
            correct_letters.append(guess)
        elif letter in correct_letters:
            display += letter
        else:
            display += "_"

            print(display) 
        
    if " _ " not in chosen_word:
         lives -=1
         print(f"You lose a life for wrong guess -- {guess}")
         if lives ==0:
             game_over = True
             print(f"You lose---- It was {chosen_word}")
    
    
    
    if " _ " not in (display):
         game_over = True
     
         print("You Win !!")


    print(stages[lives])
    
```
