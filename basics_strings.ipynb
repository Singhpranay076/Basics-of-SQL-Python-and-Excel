```python
#int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)
#float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)
#str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals

x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3

x = float(1)     # x will be 1.0
y = float(2.8)   # y will be 2.8
z = float("3")   # z will be 3.0
w = float("4.2") # w will be 4.2

x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0'
```

```python
#sub-scripting

print("Pranay Singh"[5])
print("Pranay Singh"[-4]) #counts backwards from the end
```

```python
#Slicing Strings
#[start : stop]  start → included   stop → excluded

b = "Pranay Singh"
print(b[2:5])

b = "Pranay, Singh!"
print(b[:7])

c = "Pranay, Singh!"
print(c[8:])

```

```python
a = "Hello, World!"
print(a.upper())

b = "Hello, World!"
print(a.lower())

c = "     Hello, World!       "
print(c.strip()) # returns "Hello, World!" -- removes spaces before/after ""

d= "    Pranay     "
print(d)

```

```python
#F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

#To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} 
#as placeholders for variables and other operations.

age = 36
txt = f"My name is John, I am {age}"
print(txt)

```

```python
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
```

```python
txt = f"The price is {20 * 59} dollars"
print(txt)
```
